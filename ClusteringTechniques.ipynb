```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=400, centers=4, cluster_std=2, random_state=10)

plt.scatter(X[:,0], X[:,1], c=y_true,cmap='viridis')
ax =plt.gca()
ax.axis('equal')
```

Lets run k-means on the above dataset.

```
from sklearn.cluster import KMeans
kmeans=KMeans(4)
labels= kmeans.fit(X).predict(X)

plt.scatter(X[:,0], X[:,1], c=labels,cmap='viridis')
ax =plt.gca()
ax.axis('equal')
```

```
rng = np.random.RandomState(13)
X_stretch = np.dot(X, rng.randn(2,2))
plt.scatter(X_stretch[:,0], X_stretch[:,1], c=labels,cmap='viridis')
ax =plt.gca()
ax.axis('equal')
```

```
from sklearn.cluster import KMeans
kmeans=KMeans(4)
labels= kmeans.fit(X_stretch).predict(X_stretch)

plt.scatter(X_stretch[:,0], X_stretch[:,1], c=labels,cmap='viridis')
ax =plt.gca()
ax.axis('equal')
```

We can fix this using Expectation Maximization

```
from sklearn.mixture import GaussianMixture as GMM
gmm = GMM(n_components=4)
labels=gmm.fit(X).predict(X)

plt.scatter(X[:,0], X[:,1], c=labels,cmap='viridis')
ax =plt.gca()
ax.axis('equal')
```

```
probs = gmm.predict_proba(X)
print(probs.shape)
```

```
print(probs[:40].round(3))
```

```
labels=gmm.fit(X_stretch).predict(X_stretch)
plt.scatter(X_stretch[:,0], X_stretch[:,1], c=labels,cmap='viridis')
ax =plt.gca()
ax.axis('equal')
```

```
probs = gmm.predict_proba(X_stretch)
print(probs.shape)
print(probs[:40].round(3))
```

Let's apply clustering to the MNIST dataset

```
from sklearn.datasets import load_digits
digits= load_digits()
digits.data.shape
```

We try to infer class labels automatically using clustering

```
kmeans=KMeans(n_clusters=10, random_state=0)
clusters= kmeans.fit_predict(digits.data)
kmeans.cluster_centers_.shape
```

```
fig, ax = plt.subplots(2,5, figsize=(8,3))
centers= kmeans.cluster_centers_.reshape(10,8,8)
for axi, center in zip(ax.flat,centers):
  axi.set(xticks=[],yticks=[])
  axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

```
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(digits.target, clusters)
sns.heatmap(mat.T, square= True, annot=True, fmt='d', xticklabels=digits.target_names, yticklabels=digits.target_names)
plt.xlabel('True labels')
plt.ylabel('Predicted labels')
```

```
from scipy.stats import mode
corrected_clusters= np.zeros_like(clusters)
for i in range(10):
  mask=[clusters==i]
  corrected_clusters[mask]= mode(digits.target[mask])[0]

from sklearn.metrics import accuracy_score
accuracy_score(digits.target, corrected_clusters)
```

We apply a non-linear dimensionality reduction tool TSNE to map the 8X8 images to 2 dimensions.

```
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2 , init='random', random_state= 0)
digits_proj= tsne.fit_transform(digits.data)
```

```
digits_proj.shape
```

```
plt.scatter(digits_proj[:,0],digits_proj[:,1], c=digits.target, cmap='viridis')
```

Let's try to reduce the dimensions using a linear technique, PCA

```
from sklearn.decomposition import PCA
pca= PCA(n_components=2, svd_solver='randomized', whiten=True)
pca.fit(digits.data)
Z = pca.transform(digits.data)
```

```
plt.scatter(Z[:,0],Z[:,1], c=digits.target)
```

The non-linear technique TSNE gives better results. Thus, unsupervised learning if used appropriately can provide amazing accuracies.

```
kmeans=KMeans(n_clusters=10, random_state=0)
clusters= kmeans.fit_predict(digits_proj)
for i in range(10):
  mask=[clusters==i]
  corrected_clusters[mask]= mode(digits.target[mask])[0]

from sklearn.metrics import accuracy_score
accuracy_score(digits.target, corrected_clusters)
```
