```
#Importing Basic Libraries
import pandas as pd
import numpy as np
from numpy import linalg as LA
```

```
#Generating a 3-dimensional standard multivariate distribution
mean = [0, 0 ,0]
cov = [[1, 0.8, 0.9], [0.8, 1,0.7], [0.9, 0.7,1]]
import matplotlib.pyplot as plt
x, y, z = np.random.multivariate_normal(mean, cov, 10000).T
```

```
#Generating a 3D plot of the Gaussian distribution
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(x, y, z, 'gray')
```

```
#Generating a histogram of the Euclidean distances
dist=[]
for i in range(10000):
  dist.append(np.sqrt((x[i]-mean[0])**2 + (y[i]-mean[1])**2 + (z[i]-mean[2])**2))
density=False
plt.hist(dist,density=density,bins=None);
```

```
#The average and standard deviation of the  euclidean distances
from numpy import average, std
print(average(dist))
print(std(dist))
```

```
#Function to generate the covariance matrix
def cov(d):
  a=np.random.rand(d,d)
  ans= (a*a.T)/2
  for i in range(d):
    ans[i,i]=1
  return ans
```

```
#Function to calculate the euclidean distance
def euclidian_dist(a):
  mylist=[]
  for i in range(a.shape[0]):
    ans=LA.norm(a[i][:])
    mylist.append(ans)
  arrans=np.asarray(mylist)
  return arrans  
```
