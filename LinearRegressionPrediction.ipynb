```python
#Loading the train dataset into a pandas dataframe
import pandas as pd
names =[
't', # Time (secs)
'q1', 'q2', 'q3', # Joint angle
'dq1', 'dq2', 'dq3', # Joint velocity
'I1', 'I2', 'I3', # Motor current (A)
'eps21', 'eps22', 'eps31', 'eps32', # Strain measurements
'ddq1', 'ddq2', 'ddq3' # Joint accelerations
]
df = pd.read_csv('exp_train.csv', header=None,sep=',',
names=names, index_col=0)
df.head(6)
```

```python
#Import Libraries
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X = df[['q2','dq2','eps21','eps22', 'eps31', 'eps32','ddq2']]
```

```python
X.head(5)
```

```python
y = df['I2']
```

```python
#Fitting the linear regression model to the training dataset
model = LinearRegression().fit(X, y)
```

```python
#Calculating the root mean square error
from sklearn.metrics import mean_squared_error
y_pred1 = model.predict(X)
mean_squared_error(y, y_pred1)
```

```python
#Calculating the slope and intercept of the line
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```

```python
#Loading the test data into a pandas dataframe
names =[
't', # Time (secs)
'q1', 'q2', 'q3', # Joint angle
'dq1', 'dq2', 'dq3', # Joint velocity
'I1', 'I2', 'I3', # Motor current (A)
'eps21', 'eps22', 'eps31', 'eps32', # Strain measurements
'ddq1', 'ddq2', 'ddq3' # Joint accelerations
]
dftest = pd.read_csv('exp_test.csv', header=None,sep=',',
names=names, index_col=0)
dftest.head(6)
```

```python
df1_test=dftest[['q2','dq2','eps21','eps22', 'eps31', 'eps32','ddq2']]
```

```python
y_test=dftest[['I2']]
```

```python
y_pred2 = model.predict(df1_test)
```

```python
#Calculating the MSE on the test function
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred2)
```
